```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
df = pd.read_csv("../files/canadian_immegration_data.csv")
df.head()
```

```python
df.info()
```

```python
df.set_index('Country', inplace=True)
```

```python
df
```

```python
anos = list(map(str, range(1980, 2014)))
```

```python
anos
```

```python
brasil = df.loc["Brazil", anos]
```

```python
brasil
```

```python
brasil_dict = {
    "ano": brasil.index.tolist(),
    "imigrantes": brasil.values.tolist()
}
dados_brasil = pd.DataFrame(brasil_dict)
```

```python
dados_brasil
```

```python
plt.figure(figsize=(8,4)) # escolhendo o tamanho ideal do grafico em polegadas
plt.plot(dados_brasil['ano'], dados_brasil['imigrantes']) # plotando o grafico
plt.title("ImigraÃ§Ã£o do Brasil para o CanadÃ¡") # titulo
plt.xlabel("Ano") # rotulo do X
plt.ylabel("NÃºmero de imigrantes") # rotulo do Y
plt.xticks(['1980', '1985', '1990', '1995', '2000', '2005', '2010']) # modificando os valores do X para terem uma experiencia melhor
# plt.yticks([500, 1000, 1500, 2000, 2500, 3000])
plt.show() 
```

```python
fig, ax = plt.subplots(figsize=(8,4))
ax.plot(dados_brasil['ano'], dados_brasil['imigrantes'])
ax.set_title("ImigraÃ§Ã£o do Brasil para o CanadÃ¡\n1980 a 2013")
ax.set_xlabel("Ano")
ax.set_ylabel("NÃºmero de imigrantes")
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
plt.show()
```

```python
fig, axs = plt.subplots(1, 2, figsize=(15,5))

axs[0].plot(dados_brasil['ano'], dados_brasil['imigrantes'])
axs[0].set_title("ImigraÃ§Ã£o do Brasil para o CanadÃ¡\n1980 a 2013")
axs[0].set_xlabel("Ano")
axs[0].set_ylabel("NÃºmero de imigrantes")
axs[0].xaxis.set_major_locator(plt.MultipleLocator(5))
axs[0].grid()

axs[1].boxplot(dados_brasil['imigrantes'])
axs[1].set_title("Boxplot da imigraÃ§Ã£o do Brasil para o CanadÃ¡\n1980 a 2013")
axs[1].set_xlabel("Brasil")
axs[1].set_ylabel("NÃºmero de imigrantes")
axs[1].grid()

plt.show()
```

```python
fig, axs = plt.subplots(2, 2, figsize=(10,6))
fig.subplots_adjust(hspace=0.5, wspace=0.3)
fig.suptitle("ImigraÃ§Ã£o dos quatro maiores paÃ­ses da AmÃ©rica do Sul para o CanadÃ¡ de 1980 a 2013")

axs[0, 0].plot(df.loc['Brazil', anos])
axs[0, 0].set_title("Brasil")

axs[0, 1].plot(df.loc['Colombia', anos])
axs[0, 1].set_title("Colombia")

axs[1, 0].plot(df.loc['Argentina', anos])
axs[1, 0].set_title("Argentina")

axs[1, 1].plot(df.loc['Peru', anos])
axs[1, 1].set_title("Peru")

for ax in axs.flat:
    ax.xaxis.set_major_locator(plt.MultipleLocator(5))    

for ax in axs.flat:
    ax.set_xlabel('Ano')
    ax.set_ylabel('Imigrantes')

ymin = 0
ymax = 7000

for ax in axs.ravel():
    ax.set_ylim(ymin, ymax)

plt.show()
```

```python
america_sul = df.query("Region=='South America'")
america_sul
```

```python
america_sul_sorted = america_sul.sort_values('Total', ascending=True)
```

```python
cores = []
for Country in america_sul_sorted.index:
    if Country == 'Brazil':
        cores.append('green')
    else:
        cores.append('silver')

fig, ax = plt.subplots(figsize=(12,5))
ax.barh(america_sul_sorted.index, america_sul_sorted['Total'], color=cores)
ax.set_title("ImigraÃ§Ã£o da AmÃ©rica do Sul para o CanadÃ¡\n1980 a 2013", loc='left', fontsize='18')
ax.set_xlabel("NÃºmero de imigrantes", fontsize=14)
ax.set_ylabel("")
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)

for i, v in enumerate(america_sul_sorted['Total']):
    ax.text(v + 20, i, str(v), color='black', fontsize=10, ha='left', va='center')

ax.set_frame_on(False)
ax.get_xaxis().set_visible(False)
ax.tick_params(axis='both', which='both', length=0)

fig.savefig('imigraÃ§Ã£o_america_sul.png', transparent=False, dpi=300, bbox_inches='tight')

plt.show()
```

```python
def gerar_grafico_paleta(palette):
    fig, ax = plt.subplots(figsize=(8,4))
    ax = sns.barplot(data=top_10, y=top_10.index, x='Total', orient='h', palette=palette, hue=top_10.index, legend=False)
    ax.set_title('PaÃ­ses com maior imigraÃ§Ã£o para o CanadÃ¡\n1980 a 2013', loc='left', fontsize=16)
    ax.set_xlabel('NÃºmero de imigrantes', fontsize=14)
    ax.set_ylabel('')

    plt.show()
```
